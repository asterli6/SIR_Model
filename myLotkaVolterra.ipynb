```python
import numpy as np
import matplotlib . pyplot as plt
from scipy . integrate import solve_ivp
```

now for our parameters ;-)

```python
r1 = -3
r2 = 1.5
a11 = 0
a12 = 1
a21 = -1
a22 = 0
```

Time interval t


```python
t_initial = 0
t_final = 15

x0 = 10
y0 = 5
```

lets add the actual function

```python
def lotka_volterra(t, z, r1, r2, a11, a12, a21, a22):
  x, y = z
  return [x * (r1 - a11 *x + a12 * y), y * (r2 + a21 * x - a22 * y)]
```

```python
Y0 = [ x0 , y0 ]
t_span = ( t_initial , t_final )
sol = solve_ivp(lotka_volterra , [0 ,15] , [10 , 5] , args = ( r1 , r2 , a11 ,
a12 , a21 , a22 ) , dense_output = True )
```

Generate time array

```python
t = np . linspace ( t_initial , t_final , 300)
z = sol . sol ( t )
```

Code for plotting graph

```python
plt.plot (t , z.T )
plt.xlabel ('t')
plt.legend ([r'$x_1$ ', r'$x_2$ '] , shadow = True )
plt.title ('My Lotka - Volterra System ')
plt.show ()
```
